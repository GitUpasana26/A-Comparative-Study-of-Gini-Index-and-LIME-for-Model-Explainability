```python
import matplotlib.pyplot as plt
import numpy as np

# RF Gini Index (Global Importance)
gini_features = [
    "Glucose", "BMI", "Age", "DiabetesPedigreeFunction",
    "BloodPressure", "Pregnancies", "Insulin", "SkinThickness"
]

gini_values = [
    0.2552, 0.1671, 0.1367, 0.1235,
    0.0903, 0.0850, 0.0754, 0.0669
]

# LIME Feature Importance (Local)
lime_features = [
    "Age", "DiabetesPedigreeFunction", "BMI", "Glucose",
    "BloodPressure", "Pregnancies", "SkinThickness", "Insulin"
]

lime_values = [
    0.0654, -0.0639, 0.0544, 0.0347,
    -0.0241, -0.0122, -0.0049, 0.004
]

# Convert features to numbers for x-axis
x_gini = np.arange(len(gini_features))
x_lime = np.arange(len(lime_features))

plt.figure(figsize=(12, 8))

# Plot Gini Index Line
plt.plot(x_gini, gini_values, marker='o', linewidth=2, label="Random Forest Gini Index")

# Plot LIME Line
plt.plot(x_lime, lime_values, marker='s', linewidth=2, label="LIME Feature Contribution")

# X-axis labels
plt.xticks(x_gini, gini_features, rotation=45)

plt.title("Line Graph: Random Forest Gini Index vs LIME Feature Contributions")
plt.xlabel("Features")
plt.ylabel("Importance / Contribution")
plt.grid(linestyle="--", alpha=0.4)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 1 Your RF Gini Importance
gini_scores = {
    "Glucose": 0.2552,
    "BMI": 0.1671,
    "Age": 0.1367,
    "DiabetesPedigreeFunction": 0.1235,
    "BloodPressure": 0.0903,
    "Pregnancies": 0.0850,
    "Insulin": 0.0754,
    "SkinThickness": 0.0669
}

# 2 Your LIME Global Importance
lime_scores = {
    "Glucose": 0.253373,
    "BMI": 0.154318,
    "Age": 0.081258,
    "DiabetesPedigreeFunction": 0.076716,
    "Pregnancies": 0.061378,
    "BloodPressure": 0.021342,
    "SkinThickness": 0.014891,
    "Insulin": 0.014321
}

# Convert to DataFrame
df = pd.DataFrame({
    "Feature": list(gini_scores.keys()),
    "Gini": list(gini_scores.values()),
    "LIME": [lime_scores[f] for f in gini_scores.keys()]
})

# Sort by Gini (optional)
df = df.sort_values(by="Gini", ascending=False)

# X positions
x = np.arange(len(df["Feature"]))

plt.figure(figsize=(12, 6))

# Plot as lines
plt.plot(x, df["Gini"], marker='o', linewidth=2, label="Gini Importance")
plt.plot(x, df["LIME"], marker='s', linewidth=2, label="LIME Global Importance")

# Feature labels on x-axis
plt.xticks(x, df["Feature"], rotation=45, ha='right')

plt.ylabel("Importance Score")
plt.title("Feature Importance Comparison : Gini vs LIME Global")
plt.grid(linestyle="--", alpha=0.3)
plt.legend()
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# -------------------------
# Data
# -------------------------
features = ["Glucose", "BMI", "Age", "DiabetesPedigreeFunction",
            "BloodPressure", "Pregnancies", "Insulin", "SkinThickness"]

gini = [0.2552, 0.1671, 0.1367, 0.1235, 0.0903, 0.0850, 0.0754, 0.0669]
lime_local = [0.0334, -0.1848, 0.0665, 0.0889, -0.0076, 0.0994, -0.0430, -0.0256]
lime_global = [0.253868, 0.153898, 0.081142, 0.077085,
               0.021050, 0.061295, 0.014139, 0.015056]

# -------------------------
# Line Graph
# -------------------------
x = np.arange(len(features))

plt.figure(figsize=(14, 6))

plt.plot(x, gini, marker='o', linewidth=2, label="Gini Importance")
plt.plot(x, lime_local, marker='s', linewidth=2, label="LIME Local")
plt.plot(x, lime_global, marker='^', linewidth=2, label="LIME Global")

plt.xticks(x, features, rotation=45, ha='right')
plt.ylabel("Importance Score")
plt.title("Feature Importance (Line Graph): Gini vs LIME Local vs LIME Global")
plt.grid(linestyle='--', alpha=0.3)
plt.legend()
plt.tight_layout()

plt.show()

```
